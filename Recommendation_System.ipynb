```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
import random
import warnings
warnings.filterwarnings('ignore')
```

```python
ratings = pd.read_excel(r'/content/drive/MyDrive/response to stress.xlsx', index_col=0)
ratings = ratings.fillna(0)
ratings
```

```python
def standardize(row):
    new_row = (row - row.mean()) / (row.max() - row.min())
    return new_row

ratings_std = ratings.apply(standardize)
ratings_std
```

```python
# Transpose to get similarity between users in the same rows
user_similarity = cosine_similarity(ratings_std)
user_similarity
```

```python
user_similarity_df = pd.DataFrame(user_similarity, index=ratings.index, columns=ratings.index)
user_similarity_df
```

```python
# make recommendations

stress_value = random.randint(1, 5) 
def get_stress_reduction_method(user, stress_value):
    mean_rating = 2.5
    similar_score = user_similarity_df[user] * (stress_value - mean_rating) # 2.5 is the mean of the ratings
    similar_score = similar_score.sort_values(ascending=False)
    return similar_score[:10]

num = random.randint(1, 38)
user = f"User {num}"
print(user)
similar_users = get_stress_reduction_method(user, stress_value)
users = similar_users.index.values.tolist()
print(users)
```

```python
new_ratings = ratings.loc[users, :]
new_ratings = new_ratings.fillna(0)
new_ratings
```

```python
def standardize(row):
    new_row = (row - row.mean()) / (row.max() - row.min())
    return new_row

new_ratings_std = new_ratings.apply(standardize)
new_ratings_std
```

```python
nan_list = new_ratings_std.columns[new_ratings_std.isna().any()].tolist()
for method in nan_list:
    new_ratings_std.drop([method], axis = 1)

for method in nan_list:
    new_ratings_std[method] = new_ratings_std.mean(axis=1)
```

```python
# taking transpose to get similarity between items in the same rows
item_similarity = cosine_similarity(new_ratings_std.T)
print(item_similarity)
```

```python
item_similarity_df = pd.DataFrame(item_similarity, index=ratings.columns, columns=ratings.columns)
item_similarity_df
```

```python
# pearson correlation
"""item_similarity_df = user_ratings.corr(method='pearson')
print(item_similarity_df.head(50))"""
```

```python
# make recommendations

def get_stress_reduction_recommendation(method, rating):
    mean_rating = 2.5
    similar_score = item_similarity_df[method] * (rating - mean_rating)
    similar_score = similar_score.sort_values(ascending=False)

    return similar_score

print(get_stress_reduction_recommendation("Sleep well", 4))
```

```python
# stress_reduction_user = [("Taking Yoghurt", 4), ("Taking Green Tea supplement", 3), ("Yoga", 5)]
stress_reduction_user = [('Drinking 3 litres of water daily', 3), ('Taking a walk', 2), ('laughter', 2), ('Meditation', 3), 
                         ('Avoid Tobacco and nicotine', 2), ('Avoiding Procrastination', 4), ('Deep Breating', 5), ('Yoga', 5), ('Taking a break to relax', 2), 
                         ('Listening to music', 4), ('Taking Yoghurt', 2), ('Eating Eggs', 3), ('Taking Green Tea supplement', 2), ('Eating Turkey', 3), 
                         ('Taking Oatmeal', 4), ('Eating Vegetables', 2), ('Eating Citrus Fruits like Oranges and Grapefruit', 5), ('Prayer', 5), 
                         ('Taking time off', 4), ('Sleep well', 2), ('Avoiding Caffeine', 4), ("Don't work more than 7 hours", 4), ('Avoid Negativity', 4)]

similar_method = pd.DataFrame()

for method, rating in stress_reduction_user:
    similar_method = similar_method.append(get_stress_reduction_recommendation(method, rating), ignore_index=True)

similar_method.head()
```
